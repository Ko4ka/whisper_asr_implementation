```python
import os
import requests

def simplify_transcript(data):
    simplified = {}
    for file, speakers in data.items():
        simplified[file] = {}
        for speaker, segments in speakers.items():
            simplified[file][speaker] = [
                {
                    "start": segment["start"],
                    "end": segment["end"],
                    "text": segment["text"].strip()
                } for segment in segments
            ]
    return simplified

# Directory with all audio files
audio_dir = 'C:/Users/Alex/Desktop/–ü–ª–∞–Ω–µ—Ç–∞ –ê–≤—Ç–æ/–í—Å–µ –∑–≤–æ–Ω–∫–∏/'
all_files = [os.path.join(audio_dir, f) for f in os.listdir(audio_dir) if f.endswith(('.mp3', '.wav'))]

# Split into batches of 2
def batch_files(files, n=2):
    for i in range(0, len(files), n):
        yield files[i:i + n]

url = "http://127.0.0.1:8000/transcribe_audio_local"
ALL_AUDIO = []

for batch in batch_files(all_files, 2):
    print(f"Transcribing batch: {[os.path.basename(p) for p in batch]}")
    languages = ['ru'] * len(batch)
    print(languages)
    response = requests.post(
        url,
        headers={},
        params={"languages": ",".join(languages)},
        json=batch
    )
    
    if response.ok:
        simplified = simplify_transcript(response.json())
        ALL_AUDIO.append(simplified)
        print("‚úÖ ASR OK")
    else:
        print("‚ùå ASR failed:", response.status_code, response.text)

print("üîÅ Done transcribing all files.")
```

```python
checkwords = ["–º–æ–¥–æ ", "–æ–º–æ–¥–∞", "omoda"]  # your list of words
IDS = []

for batch in ALL_AUDIO:

    for call_id, speakers in batch.items():
        found = False
        for speaker_segments in speakers.values():  # Speaker 0, Speaker 1
            for segment in speaker_segments:
                text = segment["text"].lower()
                #print(text)
                if any(word in text for word in checkwords):
                    IDS.append(call_id)
                    found = True
                    break  # stop checking this speaker
            if found:
                break  # stop checking other speakers
```
